```python
# Importing dependencies

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ev_cars = pd.read_csv('EVIndia.csv')
```

```python
ev_cars.head()
```

```python
ev_cars.describe()
```

```python
ev_cars.info()
```

```python
ev_cars.isnull().sum()
```

```python
ax = sns.countplot(data=ev_cars, x='Style')
plt.title("Count of cars based on Body Type")
plt.xlabel("Body Type")
plt.xticks(rotation='vertical')
plt.ylabel("Count")
for label in ax.containers:
    ax.bar_label(label)
plt.show()
```

```python
ax = sns.countplot(data=ev_cars, x='PriceRange')
plt.title("Count of cars based on Price Range")
plt.xlabel("Price")
plt.xticks(rotation='vertical')
plt.ylabel("Count")
for label in ax.containers:
    ax.bar_label(label)
plt.show()
```

```python
ax = sns.countplot(data=ev_cars, x='Capacity')
plt.title("Count of cars based on Seating Capacity")
plt.xlabel("Seating Capacity")
plt.xticks(rotation=45)
plt.ylabel("Count")
for label in ax.containers:
    ax.bar_label(label)
plt.show()
```

```python
ax = sns.countplot(data=ev_cars, x='BootSpace')
plt.title("Count of cars based on Boot Space")
plt.xlabel("Boot Space")
plt.xticks(rotation = 45)
plt.ylabel("Count")
for label in ax.containers:
    ax.bar_label(label)
plt.show()
```

This dataset can be found __[here](https://www.kaggle.com/datasets/kkhandekar/electric-vehicles-india)__

Most electric vehicles in India are of `Compact SUV` body type which helps us to know that if the company makes their EVs in this same category, there can be some competition. But in other category like hatchbacks, or premium hatchbacks, monopoly in the market _can be achieved_.

From this dataset, price range does not give much information. Another to be noted is that most EVs in India have `5 seats`. Therefore, if a company launches a new EV with more number of seats, it can be profitable for that company. 

From the Boot Space graph, it is noted that there are very few EVs in the `0L - 300L` range. Most of them are in `400L - 600L` range.
